```python
import efl_mapping as efl
```

```python
origin = "NG13AA"
stadium_list = {}

df = efl.get_efl_info()

for stadium, club in zip(df.Stadium, df.Club):
    print(stadium, club)
    destination = stadium + club
    destination = efl.replace_space_with_plus(destination)
    directions = efl.get_directions(origin, destination)
    distance = efl.get_distance(directions)
    lat_long = efl.get_lat_long(directions)
    stadium_list[stadium] = {
        "distance": efl.convert_distance_m_to_km(distance),
        "latitude": lat_long[0],
        "longitude": lat_long[1],
    }
    
lat_long_df = pd.DataFrame.from_dict(stadium_list).T
df = df.merge(lat_long_df, how='inner', left_on='Stadium', right_index=True)
print(df)
```
